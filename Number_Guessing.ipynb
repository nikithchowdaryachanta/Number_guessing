```python
#Importing Random Library
import random
#Asking For lower limit from the user
lowerbound=int(input("Lower Bound:- "))
#Asking for upper limit from the user
upperbound=int(input("Upper Bound:- "))
#Generating Random integer in between the limit given by user
r = random.randint(lowerbound,upperbound)
print("   You've only 7 chances to guess the Integer.")
print("   Number Range is ["+str(lowerbound)+","+str(upperbound)+"]")
print("   Now start guessing the number")
# Check the number is Even or Not
even=r%2
# Check the number is a divisible of 3 or Not
three=r%3
# Check the number is a divisible of 4 or Not
four=r%4
# Check the number is a divisible of 5 or Not
five=r%5
# Check the number is a divisible of 7 or Not
seven=r%7
# Function to Check the number is a Prime or Not
if r > 1:
    for i in range(2, int(r / 2) + 1):
        if (r % i) == 0:
            prime=1
            break
    else:
        prime=0
else:
    prime=1
#For loop to guess number Multiple times if the previous guess is wrong
for i in range(0,7):   # 0,7 to give only 7 Chances to user to guess
    if i==6:
        print("\n   *** This is your last chance to guess. Be Carefull! ***")
        print("*** Consider all the given Hints and then guess the number ***")
    guess = int(input("\nGuess the number:- "))
    # If Number is Guessed then break the loop
    if guess==r:
        print("Congratulations! Your Guess is Correct.")
        print("You guessed in "+ str((i+1))+" attempt.\n")
        #Calculating Score
        score=((7-i)/7)*100
        print("Your Score Percentage is "+str(score)+"% .")
        break;
    elif guess<r or guess>r:
        if guess<r:
            print("Try Again! You Guessed too small")
        if guess>r:
            print("Try Again! You Guessed too high")
        #Giving Hints to user to guess the number
        if i==0:
            if even==0:
                print("Hint:- Number is Even")
            else:
                print("Hint:- Number is Odd")
        if i==1:
            if three==0:
                print("Hint:- Number is a divisible by 3")
            else:
                print("Hint:- Number is not divisible by 3")
        if i==2:
            if four==0:
                print("Hint:- Number is a multiple of 4")
            else:
                print("Hint:- Number is not the multiple of 4")
        if i==3:
            if five==0:
                print("Hint:- Number is a divisible by 5")
            else:
                print("Hint:- Number is not divisible by 5")
        if i==4:
            if seven==0:
                print("Hint:- Number is a multiple of 7")
            else:
                print("Hint:- Number is not the multiple of 7")
        if i==5:
            if prime==0:
                print("Hint:- Number is a Prime Number")
            else:
                print("Hint:- Number is not a Prime Number")
        # Final Chance to Guess the number
        if i==6:
            print("Ohho! Even after so many hints, you could not guess the number.")
            print("Your Score Percentage is 0% . You are Fail")
            print("Correct Number was "+str(r)+".")
            print("Better Luck Next Time!\n")
```
